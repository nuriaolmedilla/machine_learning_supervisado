```python
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
from sklearn.tree import DecisionTreeClassifier
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 1. Cargar el dataset y dividir los datos
from sklearn.datasets import load_iris
data = load_iris()
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.3, random_state=42)
```

```python
# 2. Entrenar el modelo
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
# 3. Calcular la matriz de confusión en valores absolutos
conf_matrix_abs = confusion_matrix(y_test, y_pred)
print("Matriz de confusión (valores absolutos):\n", conf_matrix_abs)
```

```python
# Graficar la matriz de confusión con valores absolutos
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_abs, annot=True, fmt='d', cmap='Blues', xticklabels=data.target_names, yticklabels=data.target_names)
plt.title("Matriz de Confusión (Valores Absolutos)")
plt.xlabel("Predicción")
plt.ylabel("Valor Verdadero")
plt.show()
```

```python
# 4. Calcular la matriz de confusión en valores relativos
conf_matrix_rel = conf_matrix_abs / conf_matrix_abs.sum(axis=1)[:, np.newaxis]
print("Matriz de confusión (valores relativos):\n", conf_matrix_rel)
```

```python
# Graficar la matriz de confusión con valores relativos
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_rel, annot=True, fmt='.2%', cmap='Blues', xticklabels=data.target_names, yticklabels=data.target_names)
plt.title("Matriz de Confusión (Valores Relativos)")
plt.xlabel("Predicción")
plt.ylabel("Valor Verdadero")
plt.show()
```

```python
# 5. Calcular métricas
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
specificity = np.diag(conf_matrix_abs) / np.sum(conf_matrix_abs, axis=1)
```

```python
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"Specificidad: {specificity.mean():.2f}")
```
